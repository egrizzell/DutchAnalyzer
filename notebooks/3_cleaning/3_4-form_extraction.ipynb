##### Verb/Werkwoord Tables/Lexeme
Types of verbs by conjugation rules:
- Weak (de/te)
- Strong (Classes 1-7) <br>
Class 1: pattern ij-ee-ee <br>
Class 2: pattern ie-oo-oo or ui-oo-oo <br>
Class 3: pattern i-o-o or e-o-o <br>
Class 4: pattern ee-a/aa-oo <br>
Class 5: pattern ee-a/aa-ee or i-a/aa-ee <br>
Class 6: pattern aa-oe-aa <br>
Class 7: pattern X-ie-X (specifically, oo-ie-oo, a-ie-a, a-i-a, ou-iel-ou, aa-ie-aa or oe-ie-oe) <br>
Other strong verbs, which do not follow any of the above patterns <br>
- Mixed (past is weak simple/strong perfect or strong simple/weak perfect)
- Irregular
- Seperable 

Verb Tenses
- Present/tegenwoordige tijd (Onvoltooid Tegenwoordige Tijd (OTT) Simple/ Voltooid Tegenwoordige Tijd (VTT)/Perfect) 
- Past/verleden tijd (Onvoltooid Verleden Tijd (OVT) Simple/ Voltooid Verleden Tijd (VVT) Perfect)
- Future/toekomende tijd (Simple onvoltooid tegenwoordige (OTTT) /Perfect (VTTT))
- Future Conditional (OVTT/VVTT)
- Continuous (Simple/Perfect)

##### Mood:
- General (indicative)
- Hypothetical/desired (subjunctive)
- Imperative (commands/orders)
- Passive

### Inflection/declension - Nouns, adjectives, adverbs, pronouns
- Nouns/zelfstandig naamwoorden: base, plural, diminuative
- Adjectives/bijvoeglijk naamwoord: base, inflected, partitive, positive, comparative, superrelative, nounified, adverbified
- Pronouns/voornaamwoorden: 1st, 2nd, 3rd person; singular/plural; formal/neuter/masculine/feminine/object(e.g. thing); subject/(nominative/nominatief)/object-voorwerp(accusative/dative, accusatief/datief)/genetive (genitief); stressed/unstressed
- Possive Pronouns/bezittelijk voornaamwoord: Also have 1st, 2nd, 3rd person; singular/plural; formal/neuter/masculine/feminine/object(e.g. thing); stressed/unstressed; plus inflected/uninflected
- Distance Related (Demonstrative) Pronouns/aanwijzend voornaamwoorden: close(Proximal demonstrative)/far(Distal demonstrative); singular/plural; masculine/feminine/neuter
- Distance Related Adverbs (Pronominal adverbs) <-> Distance Related Pronouns betrekkelijk voornaamwoord: personal/close/far/question word or relative(using who/what as a pronoun the dog who ran away)/universal(everything/everywhere) /indefinite (something/somewhere) - onbepaald voornaamwoord/negative

Dutch Translations:
neuter: onzijdig, masculine: mannelijk, feminine: vrouwelijk
